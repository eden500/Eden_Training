```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = plt.imread("D:\Projects\pythonProject\CV\course_resources\ps_0\lena.jpg")
img2 = plt.imread("D:\Projects\pythonProject\CV\course_resources\ps_0\woman.tiff")

plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
```

```python
img1_writable = img1.copy()

# Perform the swap on the writable copy
temp = img1_writable[:, :, 0].copy()
img1_writable[:, :, 0] = img1_writable[:, :, 2]
img1_writable[:, :, 2] = temp

plt.imshow(img1_writable)
plt.title("Swapped red and blue channels")
plt.show()
```

```python
monochrome_green = img1[:, :, 1]
plt.imshow(monochrome_green)
plt.title("Monochrome image green channel")
plt.show()
```

```python
monochrome_red = img1[:, :, 0]
plt.imshow(monochrome_red)
plt.title("Monochrome image red channel")
plt.show()
```

```python
monochrome_red_center = monochrome_red[monochrome_red.shape[0] // 2 - 50:monochrome_red.shape[0] // 2 + 50,
                        monochrome_red.shape[1] // 2 - 50:monochrome_red.shape[1] // 2 + 50]
monochrome_red_center.shape
```

```python
monochrome_img2 = img2[:, :, 0].copy()
monochrome_img2[monochrome_img2.shape[0] // 2 - 50: monochrome_img2.shape[0] // 2 + 50,
                   monochrome_img2.shape[1] // 2 - 50: monochrome_img2.shape[1] // 2 + 50] = monochrome_red[monochrome_red.shape[0] // 2 - 50: monochrome_red.shape[0] // 2 + 50,
                   monochrome_red.shape[1] // 2 - 50: monochrome_red.shape[1] // 2 + 50]
plt.imshow(monochrome_img2)
plt.show() 
```

```python
img1_green = monochrome_green
print(f"img1_green mean is {np.mean(img1_green)}, and std is {np.std(img1_green)}")
```

```python
img1_green_normalized = (img1_green - np.mean(img1_green)) / np.std(img1_green) * 10
plt.imshow(img1_green)
plt.show()
plt.imshow(img1_green_normalized)
plt.show()
```

```python
img1_green_shifted = np.hstack((img1_green[:, 2:], np.zeros((img1_green.shape[0], 2), dtype=img1_green.dtype)))
plt.imshow(img1_green_shifted)
plt.show()
```

```python
difference = abs(img1_green - img1_green_shifted)
plt.imshow(difference)
plt.show()
```
