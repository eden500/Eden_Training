```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('star.jpg')
plt.imshow(img)
plt.show()
```

# Clustering using KMeans

```python
from Clustering.kmeans import ImageKmeans
```

```python
seed = 40
kmeans = ImageKmeans(num_clusters=2, max_iter=100, use_distance=False, random_state=seed)
kmeans.fit(img)
kmeans_dist = ImageKmeans(num_clusters=2, max_iter=100, use_distance=True, random_state=seed, distance_scale=0.5)
kmeans_dist.fit(img)
```

```python
clustered_img = kmeans.predict(img)
clustered_img_dist = kmeans_dist.predict(img)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(clustered_img)
plt.title("Kmeans no distance")
plt.subplot(1, 2, 2)
plt.imshow(1-clustered_img_dist)
plt.title("Kmeans with distance")
plt.show()
```

# Clustering using Agglomerative Clustering

```python
from Clustering.agglomerative import ImageAgglomerative
```

```python
agglo = ImageAgglomerative(num_clusters=2, linkage='single', use_distance=False, random_state=seed, sample_size=1)
agglo.fit(img)
agglo_dist = ImageAgglomerative(num_clusters=2, linkage='single', use_distance=True, distance_scale=0.5, random_state=seed, sample_size=1)
agglo_dist.fit(img)
```

```python
clustered_img = agglo.predict(img)
# clustered_img_dist = agglo_dist.predict(img)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(clustered_img)
plt.title("Agglomerative no distance")
plt.subplot(1, 2, 2)
plt.imshow(1-clustered_img_dist)
plt.title("Agglomerative with distance")
plt.show()
```
